## Data cleaning

We can see that with the exception of the first columns, the whole dataset is composed of float numbers describing the exchange rate with the Euro. Before we start working on the visualization, we need to parse the columns stored as object to float.

```python
# the following code was provided by Dataquest 
exchange_rates.rename(columns={'[US dollar ]': 'US_dollar',
                               '[Brazilian real ]': 'BR_real',
                               'Period\\Unit:': 'Time'},
                      inplace=True)
exchange_rates['Time'] = pd.to_datetime(exchange_rates['Time'])
exchange_rates.sort_values('Time', inplace=True)
exchange_rates.reset_index(drop=True, inplace=True)
```

```python
euro_to_dollar = exchange_rates[['US_dollar', 'Time']]
euro_to_real = exchange_rates[['BR_real', 'Time']]
```

```python
euro_to_dollar['US_dollar'].value_counts()
```

```python
euro_to_real['BR_real'].value_counts()
```

```python
# dropping the rows with no real values, "-"
# converting from int to float

# eur_dollar
euro_to_dollar = euro_to_dollar[euro_to_dollar['US_dollar'] != "-"]
euro_to_dollar['US_dollar'] = euro_to_dollar['US_dollar'].astype(float)

#eur_real
euro_to_real = euro_to_real[euro_to_real['BR_real'] != "-"]
euro_to_real['BR_real'] = euro_to_real['BR_real'].astype(float)
```

## Rolling Mean

Visualizing the evolution of the exchange rates for Dollar and Real

```python
import matplotlib.pyplot as plt
import matplotlib.style as style
%matplotlib inline

plt.plot(euro_to_dollar['Time'],
         euro_to_dollar['US_dollar'])
plt.show()
```

```python
plt.plot(euro_to_real['Time'],
         euro_to_real['BR_real'])
plt.show()
```

```python
# creating a column for the rolling mean using a moving window of 30 days
euro_to_dollar['rolling_mean'] = euro_to_dollar['US_dollar'].rolling(30).mean()
```

```python
# creating a column for the rolling mean using a moving window of 30 days
euro_to_real['rolling_mean'] = euro_to_real['BR_real'].rolling(30).mean()
```
